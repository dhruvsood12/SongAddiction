# 02 — Exploratory data analysis

We compare **sticky** songs (top ~20% by popularity in this dataset) to the rest, and relate audio features to **popularity** (our public proxy for broad replayability / stickiness).

**Hypotheses (exploratory, not assumed true):**
- Higher **danceability** or **energy** may co-occur with stickier tracks.
- **Shorter duration** may co-occur with higher popularity (less fatigue, easier repeats).
- **Valence** may matter less than arousal-related features (e.g. energy).
- **Genre** may moderate these patterns.

We report what the data supports and avoid over-interpreting noise.


```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.insert(0, str(PROJECT_ROOT))

from src import data_prep
from src import visuals

FIG_DIR = PROJECT_ROOT / "outputs" / "figures"
FIG_DIR.mkdir(parents=True, exist_ok=True)

DATA_PATH = PROJECT_ROOT / "data" / "processed" / "spotify_model_data.csv"
df = pd.read_csv(DATA_PATH)
print(df.shape, list(df.columns))
```

## 1. Popularity distribution


```python
fig = visuals.plot_popularity_distribution(df, save_path=FIG_DIR / "01_popularity_histogram.png")
plt.show()
```

## 2. Sticky vs non-sticky: feature distributions


```python
feat_for_box = [c for c in data_prep.MODEL_NUMERIC_FEATURES if c in df.columns]
fig = visuals.plot_feature_boxplots(df, feat_for_box, target_col="sticky", save_path=FIG_DIR / "02_boxplots_by_sticky.png")
plt.show()
```

## 3. Feature vs popularity (scatter + trend)


```python
for name, xcol in [("danceability", "danceability"), ("energy", "energy"), ("duration", "duration_ms")]:
    if xcol not in df.columns:
        continue
    fig = visuals.plot_scatter_with_trend(df, x_col=xcol, y_col="popularity", save_path=FIG_DIR / f"03_scatter_{xcol}_popularity.png")
    plt.show()
```

## 4. Correlation analysis


```python
corr_cols = ["popularity", "popularity_z"] + [c for c in data_prep.MODEL_NUMERIC_FEATURES if c in df.columns]
corr_cols = list(dict.fromkeys(corr_cols))
fig = visuals.plot_correlation_heatmap(df, corr_cols, save_path=FIG_DIR / "04_correlation_heatmap.png")
plt.show()
```

## 5. Top 10% vs bottom 10% by popularity


```python
q_hi = df["popularity"].quantile(0.9)
q_lo = df["popularity"].quantile(0.1)
df["pop_decile_group"] = np.where(df["popularity"] >= q_hi, "top_10pct", np.where(df["popularity"] <= q_lo, "bottom_10pct", "middle"))
sub = df[df["pop_decile_group"].isin(["top_10pct", "bottom_10pct"])].copy()

feat_compare = [c for c in data_prep.MODEL_NUMERIC_FEATURES if c in sub.columns]
fig = visuals.plot_feature_mean_comparison(sub, feat_compare, group_col="pop_decile_group", save_path=FIG_DIR / "05_top_vs_bottom10_means.png")
plt.show()
```

## 6. Genre-level analysis (if `genre` is present)


```python
has_genre = "genre" in df.columns and df["genre"].notna().any()
if not has_genre:
    print("No usable genre column — skipping genre-specific plots.")
else:
    g = df.dropna(subset=["genre"]).copy()
    g["genre"] = g["genre"].astype(str).str.strip()
    top_genres = g["genre"].value_counts().head(12).index
    gg = g[g["genre"].isin(top_genres)]

    pop_by_genre = gg.groupby("genre")["popularity"].mean().sort_values(ascending=False)
    fig, ax = plt.subplots(figsize=(10, 5))
    pop_by_genre.plot(kind="bar", ax=ax, color="steelblue", edgecolor="white")
    ax.set_title("Average popularity by genre (top 12 genres by count)")
    ax.set_ylabel("Mean popularity")
    plt.xticks(rotation=35, ha="right")
    fig.tight_layout()
    fig.savefig(FIG_DIR / "06_mean_popularity_by_genre.png", bbox_inches="tight")
    plt.show()

    sticky_rate = gg.groupby("genre")["sticky"].mean().sort_values(ascending=False)
    fig, ax = plt.subplots(figsize=(10, 5))
    sticky_rate.plot(kind="bar", ax=ax, color="darkseagreen", edgecolor="white")
    ax.set_title("Sticky rate (top 20% proxy) by genre")
    ax.set_ylabel("Mean sticky label")
    plt.xticks(rotation=35, ha="right")
    fig.tight_layout()
    fig.savefig(FIG_DIR / "07_sticky_rate_by_genre.png", bbox_inches="tight")
    plt.show()

    if "danceability" in gg.columns:
        fig, ax = plt.subplots(figsize=(11, 5))
        sns.violinplot(data=gg, x="genre", y="danceability", order=top_genres[:8], ax=ax, inner="box", palette="muted")
        ax.set_title("Danceability distribution by genre (subset)")
        plt.xticks(rotation=30, ha="right")
        fig.tight_layout()
        fig.savefig(FIG_DIR / "08_violin_danceability_by_genre.png", bbox_inches="tight")
        plt.show()
```
